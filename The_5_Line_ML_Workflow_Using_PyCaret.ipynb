```python
import pandas as pd
from pycaret.classification import *
url = 'https://raw.githubusercontent.com/pycaret/datasets/refs/heads/main/data/common/juice.csv'
juice_data = pd.read_csv(url)

#the magic 5 lines starts frome here

#line 1
s = setup(data=juice_data, target='Purchase', train_size=0.7, session_id=123, normalize=True)
best_model = compare_models()  #line 2
tuned_model = tune_model(best_model) # line 3
predictions = predict_model(best_model) #line 4
metrics = pull()  # line 5
print(metrics)
```
